# Basic Terms

- Experiment
- Sample Space
- Experimental outcomes
    + probability for experimental outcomes
- Events
    + probability for events

# Examples

## Throwing a Dice

- Experiment: Throwing a Dice
- Experimental outcomes: lands on Heads or lands on Tails
- Sample Space: {Heads, Tails}
- Event: There are only two experimental outcomes. 

## Presidential Election

After various primaries there are four major candidates who could become president. Voting takes places in half a year. Over this year, many events could happen that would increase or decrease the support for candidates. After voting, the outcome is certain. But from today's perspective, when voting is one year from now, outcomes are uncertain. 

- Experiment: Presidential Election in One Year
- Experimental outcomes: candidates DT, HC, JS, GJ winning. 
- Sample Space: {DT, HC, JS, GJ}
- Event: A woman wins is an event. A man wins. Someone from a small party wins. 


## Playing the Lottery Three Times

One could win or lose a lottery. 

- Experiment: Three Lottery Plays
- Experimental outcomes: 8 sequences of outcomes
- Sample Space: 8 Elements
- Event: Win twice, win once/lose twice


# Probability Rules

## Basic Rules

- Use $E_i$ to denote different events in the sample space.
- Each element of the sample space can not have negative probability of happening, and also can not have more than $1$ probability of happening.
    + $0 \le P(E_i) \le 1$
- The probabilities of events in the sample space summed up must be equal to $1$
    + $\Sigma_{i=1}^{n} P(E_i) = 1$
